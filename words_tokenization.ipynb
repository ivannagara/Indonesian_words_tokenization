```python
import fitz  # PyMuPDF
import re
```

```python
def extract_text_from_pdf(pdf_path):
    # Open the provided PDF file
    document = fitz.open(pdf_path)
    text = ""

    # Iterate over each page in the PDF
    for page in document:
        # Extract text from the page and add it to the cumulative text variable
        text += page.get_text()

    document.close()
    return text

# Example usage
pdf_text = extract_text_from_pdf("./sources/Andrea Hirata - Laskar Pelangi (2005).pdf")
print(pdf_text)
```

```python
## Removing unecessary characters and formatting

def clean_text(text):
    # Remove non-ASCII characters
    text = re.sub(r'[^\x00-\x7F]+', ' ', text)
    # Remove any remaining special characters and digits
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # Replace multiple spaces with a single space
    text = re.sub(r'\s+', ' ', text)
    return text.strip()
```

```python
def normalize_case(text):
    return text.lower()
```

```python
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')

def remove_stopwords(text, language='indonesian'):
    stop_words = set(stopwords.words(language))
    return ' '.join([word for word in text.split() if word not in stop_words])
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

def stem_text(text):
    factory = StemmerFactory()
    stemmer = factory.create_stemmer()
    return stemmer.stem(text)
```

```python
def tokenize_text(text):
    return text.split()
```

```python
def process_text(text):
    text = clean_text(text)
    text = normalize_case(text)
    text = remove_stopwords(text)
    text = stem_text(text)
    tokens = tokenize_text(text)
    return tokens

processed_tokens = process_text(pdf_text)
print(processed_tokens)
```

```python
import nltk
from nltk.probability import FreqDist

# Assuming 'processed_tokens' is a list of all tokens
freq_distribution = FreqDist(processed_tokens)

# Most common tokens
print(freq_distribution.most_common(10))

# Tokens that appear only once
print(freq_distribution.hapaxes())
```
